```python
#Import
import pandas as pd
from pandasai import SmartDataframe
from pandasai.llm.openai import OpenAI
```

# Data Exploration with PandasAI

```python
First_prompt = "Who earns more than the others?"

sdf.chat(First_prompt)
```

```python
#Validate the output using Pandas function
data[data.SALARY == data.SALARY.max()]
```

```python
Second_prompt = "Which rows have null values?"

sdf.chat(Second_prompt)
```

```python
#Validate the output using Pandas function
data[data.isna().any(axis=1)]
```

```python
Third_prompt = "How many employees belong to department_id = 50?"

sdf.chat(Third_prompt)
```

```python
#Validate the output using Pandas function
len(data[data['DEPARTMENT_ID'] == 50])
```
